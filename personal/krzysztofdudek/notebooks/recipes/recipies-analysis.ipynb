```python
import pandas as pd
import json
from difflib import SequenceMatcher
```

```python
with open('./data/recipies.json') as json_file:
    recipes = json.load(json_file)
```

```python
ingredients = []
for recipe in recipes:
    for ingredient in recipe['ingredient_list']:
        ingredients.append([ recipe['url'], ingredient['ingredient_name'], 
                            ingredient['ingredient_amount'],
                            ingredient['ingredient_unit']])
```

```python
df = pd.DataFrame(ingredients, columns=["URL", "nazwa", "ilosc", "jednostka"])
```

```python
def is_similar(a, b):
    return SequenceMatcher(None, a, b).ratio() > 0.75
```

```python
is_similar('jajko','jajka')
```

```python
is_similar('jajko','pomidor')
```

```python
is_similar('jajko', 'jajka z wolnego wybiegu')
```

```python
df[df['nazwa'].apply(lambda x: is_similar('jajko', x))]['nazwa'].value_counts()
```

```python
df[df['nazwa'].apply(lambda x: is_similar('cebula', x))]['nazwa'].value_counts()
```

```python
df[df['nazwa'].apply(lambda x: is_similar('pomidor', x))]['nazwa'].value_counts()
```

```python
def words():
    for line in df['nazwa'].unique():
        for word in line.split():
            yield word
```

```python
word_list = [word for word in words()]
```

```python
word_set = set(word_list)
```

```python
len(word_set)
```

```python
stop_words = set([word for word in word_set if len(word) == 1])
```

```python
def remove_words(from_word_set, remove_words):
    for stop in remove_words:
        from_word_set.remove(stop)
```

```python
remove_words(word_set, stop_words)
```
