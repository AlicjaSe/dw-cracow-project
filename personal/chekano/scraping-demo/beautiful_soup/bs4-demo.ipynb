```python
def get_url(url):
    try:
        r = requests.get(url)
    except Exception as e:
        print(e)
        return None
    else:
        # do something if no exception
        print(f"Status code:  {r.status_code}")
        print(f"Content type: {r.headers['content-type']}")
        # print(r.text[:1000])
        return r.text
```

```python
def get_reciepts(bs):
    return [recept.div.a.get('href') for recept in bs.findAll('div', {'class': 'recipe'})]
```

```python
limit = 10
url = 'https://aniastarmach.pl/przepisy/'
reciepts_urls = set()
for i in range(limit):
    bs = BeautifulSoup(get_url(url), 'lxml')
    new_reciepts=get_reciepts(bs)
    print(f"Discovered {len(new_reciepts)} reciepts")
    reciepts_urls = reciepts_urls.union(new_reciepts)
    url = bs.find('div', {'class': 'button-load-more-wrapper'}).a.get('href')
len(reciepts_urls)
```
