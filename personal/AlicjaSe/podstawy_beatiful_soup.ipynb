```python
import requests
from bs4 import BeautifulSoup

import pandas as pd
```

```python
try:
    r = requests.get('https://www.sejm.gov.pl/Sejm8.nsf/agent.xsp?symbol=posglos&NrKadencji=8')
except Exception as e:
    print(e)
else:
    #do something if no exception
    print(f"Status code:  {r.status_code}")
    print(f"Content type: {r.headers['content-type']}")
    print(r.text)
    html = r.text
```

```python
bs = BeautifulSoup(html, 'html.parser')
# children vs descendants
print(bs.table)
#print(bs.html.body.h1)
#print(bs.body.h1)
#print(bs.html.h1)
```

```python
lista = bs.table.find_all('tr')
print(lista)
```

```python
ostatni_numer = None
tablica_z_glosowaniami = []

for wiersz_tabeli in lista:
    
    elementy_wiersza = wiersz_tabeli.find_all('td')
    if len(elementy_wiersza) == 3:
        if (elementy_wiersza[0].text.isdigit()):
            ostatni_numer = elementy_wiersza[0].text
            
            
        tablica = [ostatni_numer, elementy_wiersza[1].a.get('href'),
               elementy_wiersza[1].text, elementy_wiersza[2].text]

        tablica_z_glosowaniami.append(tablica)

print(tablica_z_glosowaniami)      
```

```python
df_glosowania = pd.DataFrame(tablica_z_glosowaniami, columns = ['Nr pos. Sejmu','link do głosowania', 'Data pos. Sejmu', 'Liczba głosowań'])
```
